```python
import pandas as pd
import  numpy as np
import pymongo
from pymongo import MongoClient
from time import gmtime, strftime
from datetime import datetime
```



# Current Date

```python
date = datetime.now().strftime('%Y-%m-%d')
date
```

# https://www.coronatracker.com/country/india/

```python
df = pd.read_json("https://api.coronatracker.com/v4/analytics/newcases/country?countryCode=IN&startDate=2020-04-10&endDate=2020-06-10")
```

```python
df.tail()
```

# Saving India data in csv file

```python
India = df.tail(1)
India
India.to_csv(date + "_India.csv")
```

# https://www.covid19india.org/

```python
df2 = pd.read_json("https://api.covid19india.org/updatelog/log.json")
df2 = pd.DataFrame(df2)

list = []
for i in df2["update"]:
    i = i.split()
    list.append(i)
    
```

# Maharashtra

```python
Maharashtra = [s for s in list if "Maharashtra" in s]
New_cases_in_Maharashtra = Maharashtra[0][0]
Recoveries_in_Maharashtra = Maharashtra[1][0]
Deaths_in_Maharashtra = Maharashtra[2][0]
Maharashtra = [date,New_cases_in_Maharashtra,Recoveries_in_Maharashtra,Deaths_in_Maharashtra]
Maharashtra
Maharashtra = pd.DataFrame(Maharashtra)
Maharashtra  = Maharashtra.transpose()
Maharashtra.columns = ["Date","New_cases_in_Maharashtra","Recoveries_in_Maharashtra","Deaths_in_Maharashtra"]
Maharashtra.head()
```


# Karnataka

```python
Karnataka = [s for s in list if "Karnataka" in s]
New_cases_in_Karnataka = Karnataka[0][0]
Recoveries_in_Karnataka = Karnataka[0][3]
Deaths_in_Karnataka = Karnataka[0][6]

Karnataka = [New_cases_in_Karnataka,Recoveries_in_Karnataka,Deaths_in_Karnataka]
Karnataka

Karnataka = pd.DataFrame(Karnataka)
Karnataka = Karnataka.transpose()
Karnataka.columns = ["New_cases_in_Karnataka","Recoveries_in_Karnataka","Deaths_in_Karnataka"]
Karnataka.head()
```

# Saving Maharashtra & Karnataka data in csv file

```python
df = pd.concat([Maharashtra,Karnataka],axis =1)
df
df.to_csv(date + "_State.csv",index = False)
```

# uploading state data in Mongodb

```python
client = MongoClient('localhost', 27017)
db = client.covid
state = db.state
state = pd.read_csv(date + "_State.csv") #csv file which you want to import
records_ = state.to_dict(orient = 'records')
result = db.state.insert_many(records_ )
```
